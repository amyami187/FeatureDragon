```python
import numpy as np
import sklearn as sklearn
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import random
random.seed(13)
```

```python
data,labels=make_moons(n_samples=1000, noise=.15, random_state=0)
labels = list(map(lambda x: -1 if x == 0 else x,labels))
```

```python
plt.scatter(x=data[:,0],y=data[:,1],c=labels)
```

```python
np.savez_compressed("moons.npz",data=data,labels=labels)
```
