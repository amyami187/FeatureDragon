```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sklearn
from sklearn.datasets import make_moons
random.seed(13)
```

```python
data,labels=make_moons(n_samples=1000, noise=.01, random_state=0)
labels = list(map(lambda x: -1 if x == 0 else x,labels))
```

```python
plt.scatter(x=data[:,0],y=data[:,1],c=labels)
```

```python
np.savez_compressed("moons_nonoise.npz",data=data,labels=labels)
```

```python
def normalize(data):
    data_min,data_max = min(data),max(data)
    denom =  (2*math.pi)/(data_max-data_min)
    return [ (x - data_min )*denom  for x in data]
```

```python
x_normalized = normalize(data[:,0])
y_normalized = normalize(data[:,1])
```

```python
plt.scatter(x=x_normalized,y=y_normalized,c=labels)
```

```python
data_normalized=list(zip(x_normalized,y_normalized))
```
